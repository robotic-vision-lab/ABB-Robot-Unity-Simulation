```python
import pyspacemouse
import time
```

```python
success = pyspacemouse.open(set_nonblocking_loop=True) # No callbacks here

if success:
    print("SpaceMouse connected. Reading state directly.")
    try:
        while True:
            state = pyspacemouse.read()
            # Check if movement values are non-zero, as small movements might be filtered by deadbands
            if any(abs(val) > 0.01 for val in [state.x, state.y, state.z, state.roll, state.pitch, state.yaw]):
                print(f"Translation: x={state.x:.2f}, y={state.y:.2f}, z={state.z:.2f} | "
                      f"Rotation: roll={state.roll:.2f}, pitch={state.pitch:.2f}, yaw={state.yaw:.2f}")
            time.sleep(0.01)
    except KeyboardInterrupt:
        print("Exiting.")
    finally:
        pyspacemouse.close()
else:
    print("Could not connect to SpaceMouse.")
```
